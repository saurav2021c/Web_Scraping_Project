# Web Scraping of Jobs From Indeed Website

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
#1 Starts the driver and goes to our starting webpage
driver = webdriver.Chrome()
driver.get('https://in.indeed.com/')
time.sleep(2)

#2 inputs What and Where
what = driver.find_element(By.XPATH, '//*[@id="text-input-what"]')
what.send_keys('Data Analyst')
where = driver.find_element(By.XPATH, '//*[@id="text-input-where"]')
where.send_keys('Mumbai,', 'Maharashtra')
FindButton = driver.find_element(By.XPATH, '//*[@id="jobsearch"]/button').click()
time.sleep(2)

#3 To store data in data Frame table
df = pd.DataFrame({'Links':[''], 'Company': [''], 'Job Title':[''], 'Rating':[''], 'Date':['']})

page_count = 0

#4 This loop goes through every page and grabs all the details of each posting
while True:
    #5 Imports the HTML of the current page into python
    soup = BeautifulSoup(driver.page_source, 'lxml')
    
    #6 Grabs the HTML of each posting in current page
    postings = soup.find_all('div', class_ = 'slider_container css-8xisqv eu4oa1w0')
    
    #7 grabs all the details for each posting and adds it as a row to the dataframe
    for post in postings:
        link = post.find('a', class_ = 'jcs-JobTitle css-jspxzf eu4oa1w0').get('href')
        link_full = 'https://in.indeed.com'+ link
        title = post.find('a', class_ = 'jcs-JobTitle css-jspxzf eu4oa1w0').text.strip()
        company = post.find('span', class_ = 'companyName').text
        try:
            rating = post.find('span', class_ = 'ratingNumber').text
        except:
            rating = 0
        try:
            date = post.find('span', class_ = 'date').text.strip()
        except:
            Date = 'NA'
        #8 Append the data of current page by container by container
        df = df.append({'Links':link_full, 'Company': company, 'Job Title':title, 'Rating':rating, 'Date':date}, ignore_index = True)
    
    page_count += 1
   
    #9 run till the given count of pages then it will stop the process
    if page_count>=70:
        break
       
    #10 Grab the next page url
    try:
        button = soup.find('a', {'aria-label' : 'Next Page'}).get('href')
        driver.get('https://in.indeed.com'+button)
        time.sleep(1)
    except:
        break

#-----------------------------------------------------------------------------------------------------------------#
# This is only done to sort the columns

#1 Extracting the first two characters from the 'Date' column and stripping any whitespace
df['Date_num'] = df['Date'].apply(lambda x: x[:2].strip())

#2 Function to convert the input to an integer, or return the input itself if the conversion fails
def integer(x):
    try:
        return int(x)
    except:
        return x

#3 Applying the 'integer' function to the 'Date_num' column to create 'Date_num2'
df['Date_num2'] = df['Date_num'].apply(integer)

#4 Sorting the DataFrame based on 'Date_num2' and 'Rating'
df.sort_values(by=['Date_num2', 'Rating'], inplace=True)

#5 Selecting a subset of columns and reassigning them back to the DataFrame variable 'df'
df = df[['Links', 'Company', 'Job Title', 'Rating', 'Date']]

#6 exports the data as a csv
df.to_csv('D:\\Web Scraping\\Jupyter Files\\csv_files\\indeed_jobs.csv')
```

# Code will automate generated CSV file and send it to any mail you want

```python
#1 Import the important libraries
import smtplib, ssl
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email.mime.multipart import MIMEMultipart
from email import encoders

#2 Input the email account that will send the email and who will receiving it
sender = 'ENTER_YOUR_EMAIL'
receiver = ('ENTER_EMAIL_OF_RECIEVER')

#3 Creates the Message, Subject line, From and To
msg = MIMEMultipart()
msg['Subject'] = 'New Jobs on Indeed'
msg['From'] = sender
msg['To'] = ','.join(receiver)

#4 Adds a csv file as an attachment to the email (indeed_jobs.csv is our attached csv in this case)
part = MIMEBase('application', 'octet-stream')
part.set_payload(open('D:\\Web Scraping\\Jupyter Files\\csv_files\\indeed_jobs.csv', 'rb').read())
encoders.encode_base64(part)
part.add_header('Content-Disposition', 'attachment; filename ="indeed_jobs.csv"')
msg.attach(part)

#5 This the body part in which message is attached
body_text = "Hello, this email contains information about new jobs on Indeed from Saurav."
msg.attach(MIMEText(body_text, 'plain'))


#6 Will login to your email and actually send the message above to the receiver
s = smtplib.SMTP_SSL(host = 'smtp.gmail.com', port = 465)
s.login(user = 'ENTER_YOUR_EMAIL', password = 'ENTER_YOUR_PASSWORD')
s.sendmail(sender, receiver, msg.as_string())
s.quit()
```
